```python
# 把一些警告的訊息暫時関掉
import warnings
warnings.filterwarnings('ignore')

# Utilities相關函式庫
import os

# 圖像處理/展現的相關函式庫
import matplotlib.pyplot as plt
```

```python
# 專案的根目錄路徑
ROOT_DIR = os.getcwd()

# 訓練/驗證用的資料目錄
DATA_PATH = os.path.join(ROOT_DIR, "val2018")
OUTPUT_PATH = os.path.join(ROOT_DIR, "data")
```

```python
from PIL import Image
import random
# 載入圖像檔

maxCount = 1200
count = -1

blank_image = 0

def handleImage(i1, index):  
    image = Image.open(i1)
    
    index = index % 4

    if index == 0:
        x1,y1 = (0,0)
    elif index == 1:
        x1,y1 = (0,160)
    elif index == 2:
        x1,y1 = (0,320)
    elif index == 3:
        x1,y1 = (0,480)

    x2,y2 = (x1+320, y1+160)
    bbox = (0, 0, 320, 160)
    position = (0, 160 * index)
    # 進行裁剪
    cropped_image = image.crop(bbox)
    blank_image.paste(cropped_image, position)

fileCount = -1
rN = -1
for file in os.listdir(DATA_PATH):
    count = count + 1
    if rN == -1:
        rN = random.randint(2,4)
    if (fileCount <= maxCount):
        file_path = os.path.join(DATA_PATH, file) 
        if (count == 0):
            blank_image = Image.new('RGBA', (320, rN*160), 'white')
        if (count == rN):
            handleImage(file_path, count)
            rN = -1
            count = -1
            fileCount += 1
            blank_image=blank_image.convert('RGB')
            blank_image.save(os.path.join(OUTPUT_PATH, str(fileCount) + ".jpg"))
        if (count < rN):
            handleImage(file_path, count)
    else:
        continue

```
